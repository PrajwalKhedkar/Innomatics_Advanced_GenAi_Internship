*Q4 Explain the difference between stemming and lemmatization.*  
Ans:- **Stemming
It cuts off word endings using simple rules
Does not consider meaning or grammar
May produce incorrect or non-real words
Example:
“running” → “run”
“studies” → “studi” (not a real word)
Lemmatization
Converts words to their actual root (dictionary form)
Considers meaning and grammar
Produces correct words
Example:
“running” → “run”
“better” → “good”**

Task 2:Build Advanced Preprocessing Function  
Implement the following function:  
def preprocess_text(text):  


    pass  
  
Required Features   
Your function must include:  
Remove numbers  
Example: "I have 2 dogs" → "I have dogs"  
Remove extra spaces  
Example: "This is good" → "This is good"  
Handle repeated characters  
Example: "soooo goooood!!!" → "so good"  
Convert text to lowercase  
Example: "WOW!!! This is GREAT!!!" → "wow this is great"  
Remove very short tokens     
Remove words with length ≤ 2  
Exception: keep meaningful words like "no", "not"  
Remove URLs and email-like patterns  
Example: "Visit http://example.com now" → "visit now"  


```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""
    # Lowercase
    text = text.lower()
    # Remove URLs & emails
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    # Handle repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    # Remove special characters/emojis
    text = re.sub(r'[^a-z\s]', '', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    # Tokenize
    tokens = text.split()
    # Remove short tokens (<=2) except "no", "not"
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]
    cleaned_sentence = " ".join(cleaned_tokens)
    return cleaned_tokens, cleaned_sentence

num=preprocess_text( "I have 2 dogs")
print(num)
exsp=preprocess_text("     This is good ")
print(exsp)
repchar=preprocess_text("soooo goooood!!!")
print(repchar)
lwrcase=preprocess_text("WOW!!! This is GREAT!!!")
print(lwrcase)
remlink=preprocess_text("Visit http://example.com now")
print(remlink)
```

Task 4: Token Analytics  
For each processed sentence, compute:  
Total number of tokens  
Number of unique tokens  
Average token length  
Analysis Questions  
Which sentence had the most noise?  
Which sentence retained the most meaningful tokens after cleaning?  

```python
import numpy as np

analytics = []

for tokens in results:
    if len(tokens) == 0:
        analytics.append((0, 0, 0))
        continue

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = np.mean([len(t) for t in tokens])

    analytics.append((total, unique, avg_len))

for i, (t, u, a) in enumerate(analytics):
    print(f"Sentence {i+1}: Total={t}, Unique={u}, AvgLen={a:.2f}")
```

Task 6: Build Full Pipeline  
def full_pipeline(text_list):  
    pass   
   
Expected Output  
{   
    "tokens": [...],  
    "clean_sentences": [...]   
}  

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)  
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

full_pipeline(["WOW!!! I have 2 dogs and 1 cat 😍😍. Visit https://example.com NOW!!!  Email me at test123@gmail.com. This is soooo goooood!!!  Nooooo I am NOT happy with this service 0/10!!! OK ok OK!!!"])
```
