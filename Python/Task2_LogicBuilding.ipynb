Problem 1: Unique Words in a Sentence  

*Problem Statement:*  
*Given a sentence, identify all unique words and display:*  
*The count of unique words*  
*The unique words themselves*  

*Input:*  
*Enter sentence: python is easy and python is powerful*   

*Expected Output:*  
*Unique words count: 5*     
*Unique words: {'python', 'is', 'easy', 'and', 'powerful'}* 
 
Real-Time Applications:  
Text analytics  
Natural Language Processing (NLP) preprocessing  
Search engine indexing  
Chat analysis systems  

```python
sentence="Python is easy and python is powerful"   # or input()
sentence=sentence.lower() # senetence may contain same words with different cases
words=sentence.split(" ") # splitting in words
u_words=set(words) # converting to set for finding unique words
print(u_words) # printing set 
print(len(u_words)) # length of set = unique words
```

Problem 2: Highest Salary from Employee Data 

*Problem Statement:* 
*Store employee names and their salaries, and determine which employee has the highest salary.* 

*Input Format:* 
*A dictionary where:*  
*Keys represent employee names*  
*Values represent employee salaries*  
 
*Example Input:*   
*employees = {*  
    *"Ravi": 75000,*  
    *"Anita": 68000,*  
    *"Kiran": 72000*  
*}*

*Expected Output*  
*Highest Salary: Ravi - 75000*  

Real-Time Applications:  
HR management systems  
Payroll analysis  

```python
employees = {"Anita": 68000,"Ravi": 75000,"Kiran" : 7200}
emp="" # initialize name
maxsal=0   # initialize max salary
for employee in employees:
    if(employees[employee]>maxsal): # if curr salary> max salary
        emp=employee
        maxsal=employees[employee]
print("Highest Salary:\n"+emp+"-",maxsal)
```

Problem 3: Find Maximum and Minimum Values    
   
*Problem Statement:*   
*Find the maximum and minimum values from a list of numbers.*  
   
*Input Format*  
*A list of integers where each element represents a numeric value.*  
  
*Example Input:*   
*numbers = [45, 22, 89, 10, 66]*   
   
*Expected Output:*   
*List: [45, 22, 89, 10, 66]*   
*Max: 89*   
*Min: 10*  
   
Real-Time Applications:  
Sales data analysis  
Temperature monitoring systems  
Performance metric evaluation  

```python
numbers = [45, 22, 89, 10, 66]
maxnum=minnum=numbers[0] # initialize
for num in numbers:
    if num>maxnum: # if current is greater than max
        maxnum=num
    if num<minnum: # if current is samller than min
        minnum=num
print("Numbers:",numbers)
print("Max:",maxnum)
print("Min:",minnum)
```

Problem 4: Count Products Above a Price Threshold  
  
*Problem Statement:*
*Count how many products have a price greater than 1000.*  
    
*Input Format:*  
*A list of integers where each element represents the price of a product.*  \
  
*Example Input:*  
*prices = [450, 1200, 899, 1500, 300]*  
  
*Expected Output:*   
*Products above 1000: 2*   

Real-Time Applications:  
E-commerce product filtering  
Budget analysis  

```python
prices = [450, 1200, 899, 1500, 300]
threshold=1000 # Initialize Threshold
count=0 # Initialize count
for price in prices:
    if price>1000: # count if greater than threshold
        count+=1
print("Products above Threshold:",count)
```

Problem 5: Calculate Attendance Percentage  

*Problem Statement:*
*Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).*   
  
*Example Input:*  
*attendance = ["P", "P", "A", "P", "P"]*  

*Expected Output:*  
*Attendance Percentage: 80.0*  
  
Real-Time Applications:  
Internship tracking systems  
HR attendance management systems  

```python
attendance = ["P", "P", "A", "P", "P"]
present=0
for mark in attendance:
    if mark=="P":
        present+=1
present_persentage=(present/(len(attendance)))*100
print("Attendence Percantage:",present_persentage)
```

Problem 10: Calculate Average Marks
  
*Problem Statement:*   
*Calculate the average marks of a student from a list of marks.*   
  
*Expected Output:*  
*Average Marks: 44.4*   
    
Real-Time Applications:  
Student performance evaluation   
Result processing systems   

```python
marks = [45, 42, 49, 40, 46]
sum=0
for m in marks:
    sum+=m
avg=sum/(len(marks))
print("Average Marks:",avg)
```
