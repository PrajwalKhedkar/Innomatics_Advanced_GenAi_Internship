Problem 1: Smart Light Controller  
  
Problem Statement:  
Create a class to control a smart light that can be turned ON or OFF and display its status.  
  
üåç Real-Time Use  
Smart home automation systems  
  
üì• Sample Input  
Light Name: Bedroom Light  
Action: ON  

üì§ Expected Output  
Bedroom Light is ON  

```python
class LightControl:
    def __init__(self,lights):
        self.lights=lights
        print("Welcome to SmartLight.")
        
    def check_all(self):
        for light in self.lights:
            status="ON" if self.lights[light] else "OFF"
            print(light,":",status)

    def turn_on(self,light_name):
        if self.lights[light_name]==False:
            self.lights[light_name]=True
            print(light_name,"is ON.")
        elif self.lights[light_name]==True:
            print(light_name,"is already ON")
        else:
            print("Light is Not Available.")

    def turn_off(self,light_name):
        if self.lights[light_name]==True:
            self.lights[light_name]=False
            print(light_name,"is OFF.")
        elif self.lights[light_name]==False:
            print(light_name,"is already OFF")
        else:
            print("Light is Not Available.")
    
    def actions(self,light_name,action):
        if light_name not in self.lights:
            return "Light is not Available"
        if action=="ON":
            return self.turn_on(light_name)
        elif action=="OFF":
            return self.turn_off(light_name)
        elif action=="Status":
            status="ON" if self.lights[light_name] else "OFF"
            print(light_name,":",status)
            return
        else:
            return "Wrong action."

lights={"Bedroom Light":False,"Hall Light":False,"Kitchen Light":False,"Front Yard":False,"Back Yard":False}
lc=LightControl(lights)
light_name=input("Enter Light Name: \n")
action=input("Enter Action: \n")
lc.actions(light_name,action)
lc.check_all()
```

Problem 6: Delivery Address Manager  
  
Problem Statement  
Create a class that stores customer name and delivery address and prints delivery details.  
  
Real-Time Use  
Courier and logistics platforms  
  
üì• Sample Input  
Customer Name: Suman  
Address: Hyderabad  
  
üì§ Expected Output  
Delivery Details  
Customer: Suman  
Address: Hyderabad  

```python
class Customers:
    def __init__(self):
        self.customer={}
    def add_customer(self):
        name=input("Customer Name: ")
        address=input("Address: ")
        self.customer[name]=address
    def display(self):
        for name in self.customer:
            print(f"Customer: {name}\nAddress: {self.customer[name]}")

c1=Customers()
c1.add_customer()
c1.display()
```
