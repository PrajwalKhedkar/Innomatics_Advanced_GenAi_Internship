Problem 1: Social Media – Post Engagement Analyzer   

Problem Statement:   
Create a function to analyze engagement on a social media post.   

Rules :   
Calculate total likes    
If Total Likes ≥ 1000 → Viral Post   
Otherwise → Normal Engagement   

Expected Output Format:   
Total Likes: 1050   
Post Status: Viral Post   
      
Real-Time Use : 
Social media analytics platforms   
Marketing dashboards   

```python
def analyze_engagement(likes):
    total_likes=0
    post_status=""
    for like in likes:
        total_likes+=like
    if total_likes>=1000:
        post_status+="Viral Post"
    else:
        post_status+="Normal Engagement"
    print(f"{'Total Likes':<12}:{total_likes:>12}\n{'Post Status':<12}:{post_status:>12}")

likes=[200,54,675,232,976,12]
analyze_engagement(likes)
```

Problem 2: Healthcare – Medicine Stock Alert System  

Problem Statement:   
Create a function to monitor medicine stock levels.  
  
Rules:  
Stock < 10 → Low Stock Alert   
Stock ≥ 10 → Stock Sufficient  

Expected Output Format:    
Medicine Stock: 6   
Status: Low Stock Alert  
  
Real-Time Use:  
Pharmacy management systems  
Hospital inventory tracking  

```python
def stock_check(stocks):
    stock_status=""
    for med in stocks:
        print(f"{'Medicine Stock':<15}:{med}")
        if med>=10:
            stock_status+="Sufficient Stock"
        elif med>0 and med<10:
            stock_status="Alert!! LOW STOCK"
        else:
            stock_status="Warning!!! NO STOCKS LEFT"
        print(f"{'Status':<15}:{stock_status}")
        stock_status=""
stocks=[12,4,2,14,0,17,4]
stock_check(stocks)
```

Problem 3: Agriculture – Rainfall Adequacy Checker   

Problem Statement:   
Create a function that analyzes seasonal rainfall data.  

Rules:  
Calculate average rainfall  
Average ≥ required level → Adequate Rainfall  
Otherwise → Inadequate Rainfall  

Expected Output Format:  
Average Rainfall: 72  
Rainfall Status: Adequate Rainfall  
  
Real-Time Use:  
Climate monitoring systems  
Irrigation planning tools  

```python
def rain_check(raindata,requiredlevel):
    avgrain=0
    rain_status=""
    for data in raindata:
        avgrain+=data
    avgrain=int(avgrain/(len(raindata)))
    if avgrain>=requiredlevel:
        rain_status="Adequate Rainfall"  
    else:
        rain_status="Inadequate Rainfall"
    print(f"{'Average Rainfall':<18} : {avgrain}\n{'Rainfall Status':<18} : {rain_status}")

raindata=[23,56,43,32,90,100,99,87,89,54,43,59]
required=70
rain_check(raindata,required)
```


Problem 4: Social Media – Duplicate Account Detection
  
Problem Statement:  
Create a function to detect duplicate usernames in a system.  

Rules:  
If duplicates exist → Flag accounts  
Else → Accounts are unique  
  
Expected Output Format:  
Duplicate Accounts Found: Yes  

Real-Time Use:  
Platform moderation  
Fraud detection systems  

```python
def isdupacc(user_ids):
    duplicates={}
    for user in user_ids:
        if user in duplicates:
            duplicates[user]="Yes"
        else:
            duplicates[user]="No"
    for user in duplicates:
        print(f"{user}==> Duplicate Accounts Found : {duplicates[user]}")

user_ids = ["user1", "user2", "user2", "user3", "user4", "user3"] 
isdupacc(user_ids)
```

Problem 7: System Monitoring – Application Health Checker 
   
Problem Statement: 
Create a function to evaluate application health based on error count.  
  
Rules:  
Errors = 0 → Healthy  
Errors ≤ 5 → Minor Issues   
Errors > 5 → Critical Issues   
  
Expected Output Format:   
Error Count: 7  
System Status: Critical Issues  
  
Real-Time Use:    
System monitoring tools   
IT operations dashboards   

```python
def check_system(syslogs):
    errors=syslogs.count("ERROR")
    if errors==0:
        sys_status="Healthy"
    elif errors<=5:
        sys_status="Minor Issues"
    else:
        sys_status="Critical Issues"
    print(f"Error Count: {errors}\nSystem Status: {sys_status}")
syslogs=["OK","ACCEPTED","ERROR","OK","ERROR","ERROR","ERROR"]
check_system(syslogs)
```


Problem 8: Banking – Daily Transaction Limit Checker  
  
Problem Statement:  
Create a function to validate daily transaction amount.  
  
Rules:  
Daily limit = ₹50,000  
Amount within limit → Approved  
Amount exceeds limit → Rejected  
  
Expected Output Format:  
Transaction Amount: 60000  
Transaction Status: Rejected  
  
Real-Time Use:  
Banking systems  
Fraud prevention systems   

```python
def check_transactions(transactions):
    transactions_made=0
    for transaction in transactions:
        if(transactions_made<50000):
            if(transactions_made+transaction>50000):
                print("Daily Limit is Rs. 50000")
                transaction_status="Rejected"
            else:
                transactions_made+=transaction
                transaction_status="Approved"
        else:
            transaction_status="Rejected"
        print(f"Transaction Amount: {transactions_made}\nTransaction Status: {transaction_status}")
    
transactions=[12200,3000,4000,5600,920,10999,59920,100]
check_transactions(transactions)
```

Create a function to calculate attendance percentage and check exam eligibility.

Expected Output Format:   
Attendance Percentage: 80.0

```python
def check_attendence(attendence):
    present=attendence.count("P")
    presentpercent=(present/len(attendence))*100
    if presentpercent>=75:
        eligibility_status="Eligible"
    else:
        eligibility_status="Not Eligible"
    print(f"Attendence Percentage: {presentpercent:.2f} \nExam Eligibility: {eligibility_status}")

attendence=["P","P","P","A","P","P","P","A",]
check_attendence(attendence)
```
