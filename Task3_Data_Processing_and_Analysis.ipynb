Problem Statement 1: Employee Performance Bonus Eligibility  

Description:  
A company evaluates employee performance scores at the end of the year. You are given a dictionary  containing employee names and their performance scores. 

Input:  
employees = {  
"Ravi": 92,  
"Anita": 88,  
"Kiran": 92,   
"Suresh": 85  
}  

Expected Output:  
Top Performers Eligible for Bonus: Ravi, Kiran (Score: 92)  

```python
employees = {  
"Ravi": 92,  
"Anita": 88,  
"Kiran": 92,   
"Suresh": 85  
}
max_score=max(employees.values()) # store max score 
top=[]
for emp in employees:
    if employees[emp]==max_score:
        top.append(emp)

print("Top Performers Eligible for Bonus:",", ".join(top),"(Score: "+str(max_score)+")")
```

Problem Statement 2: Search Query Keyword Analysis 

Description:   
An e-commerce website stores customer search queries. You are given a search query sentence entered by a user.  
  
Input:  
"Buy mobile phone buy phone online"  

Expected Output:  
{'buy': 2, 'phone': 2}  


```python
import re
ip="Buy mobile phone buy phone online !,2,3,425,12  "  # or ip=input()
s=ip.lower()
s=re.sub(r'[^a-zA-Z ]',"",s)
s=s.strip()
s=s.split(" ")
data={}
for w in s:
    if w in data:
        data[w]+=1
    else:
        data[w]=1
for w in s:
    if data[w]<=1:
        data.pop(w)
print(data)
```


Problem Statement 3: Sensor Data Validation  

Description:  
A factory collects sensor readings every hour. Each reading is stored in a list where the index represents the hour and the value represents the sensor reading.  
  
Input:  
sensor_readings = [3, 4, 7, 8, 10, 12, 5]  
  
Expected Output:   
Valid Sensor Readings (Hour, Value):  
[(1, 4), (3, 8), (4, 10), (5, 12)]  


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings=[]
for i in range(len(sensor_readings)):
    if sensor_readings[i]%2==0:
        valid_readings.append((i,sensor_readings[i]))
print(valid_readings)
```

Problem Statement 4: Email Domain Usage Analysis  
  
Description:  
A company wants to analyze which email providers its users are using. You are given a list of employee email IDs.  
  
Input:  
emails = [  
"ravi@gmail.com",  
"anita@yahoo.com",  
"kiran@gmail.com",  
"suresh@gmail.com",  
"meena@yahoo.com"  
]  
  
Expected Output:   
gmail.com: 60%     
yahoo.com: 40%  

```python
emails = [  
"ravi@gmail.com",  
"anita@yahoo.com",  
"kiran@gmail.com",  
"suresh@gmail.com",  
"meena@yahoo.com"  
] 
domains={}
for email in emails:
    domain=email.split("@")
    domain=domain[1]
    if domain in domains:
        domains[domain]+=1
    else:
        domains[domain]=1
for domain in domains:
    print(domain+":",(str(int(domains[domain]/len(emails)*100))+"%"))
```

Problem Statement 5: Sales Spike Detection  

Description:  
A retail company tracks daily sales. Sudden spikes in sales may indicate promotions or unusual activity.  

Input:  
sales = [1200, 1500, 900, 2200, 1400, 3000]  
  
Expected Output:   
Day 6: 3000  


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]  
avg_sales=0
for sale in sales:
    avg_sales+=sale
avg_sales=avg_sales/len(sales)
for i in range(len(sales)):
    if sales[i]>(1.3*avg_sales):
        print("Day "+str(i+1)+":",sales[i])
```

Problem Statement 6: Duplicate User ID Detection  

Description:  
A system stores user IDs during registration. Duplicate IDs can cause data integrity issues.  

Input:  
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]  

Expected Output:    
user1 → 3 times  
user3 → 2 times  

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"] 
duplicates={ }

for id in user_ids:
    if id not in duplicates:
        duplicates[id]=user_ids.count(id)

for dup in duplicates:
    if duplicates[dup]>1:
        print(dup,"->",duplicates[dup],"times")
```
