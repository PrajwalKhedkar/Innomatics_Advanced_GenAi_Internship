Problem 1: Mobile Recharge Validation System

Problem Statement:  
Create a function to validate a mobile recharge request.  
  
Rules:  
Recharge amount must be ≥ ₹50  
  
Valid plans: 199, 299, 399, 599   
User should be allowed to retry if invalid   

Real-Time Use:  
Telecom recharge apps  
Payment gateways  

```python
def recharge_validation():
    successful=False
    valid_plans=[199,299,399,599]
    while not successful:
        recharge=int(input("Enter recharge amount:"))
        if recharge<50:
            print("Amount must be more than 50 Rs.")
        elif recharge not in valid_plans:
            print("Please select valid plan(Rs. 199,Rs. 299,Rs. 399, Rs. 599)")
        else:
            print("Recharge of Rs.",recharge," is Successful!!!")
            successful=True
    
recharge_validation()
```

Problem 2: Inventory Reorder Alert System  
   
Problem Statement:  
Given a dictionary of products and stock quantities:  
Print Reorder Alert if stock < 15   
Print Stock OK otherwise  
  
Real-Time Use:  
Warehouse systems  
Retail management    


```python
def check_stock(products):
    print("Product\tStatus")
    print("-------------------------")
    for prod in products:
        if products[prod]<15:
            print(prod+"\tAlert!!!")
        else:
            print(prod+"\tStock :OK")


products={"Mouse":20,"Monitor":30,"K-board":10,"CPU":12,"HDD":25,"SSD":43}
check_stock(products)
```

Problem 3: Student Result Processing System  
   
Problem Statement:  
Create a function that:  
Accepts a list of student marks  
Calculates average marks  
Prints Pass if average ≥ 50 else Fail  

Real-Time Use: 
Online exam systems  
Evaluation portals  

```python
def result(marks):
    sum=0
    for m in marks:
        sum+=m
    average=sum/(len(marks))
    if average>=50:
        return "Pass"
    return "Fail"

marks=[76,67,87,54,89,63]
result(marks)
```

Problem 4: Cab Fare Estimator with Retry Option

Problem Statement:
Create a cab fare calculator:  
Base fare = ₹50   
₹12 per km  
Peak hour → 25% extra   
Allow retry using while loop  

Real-Time Use:  
Ride-hailing apps  
Transport systems  

```python
def calc_fare():
    flag=True
    while(flag):
        distance=float(input("Enter distance in Km :"))
        if distance<=0:
            print("Please enter valid distance !!!")
            print("Try Again !!!")
            continue
        peak_hr=input("Is this peak hour?(Y/y/N/n")
        if peak_hr not in ["y", "n"]:
            print("Please enter only Y or N")
            print("Try Again !!!")
            continue
        peak_hr=peak_hr.lower()
        base_fare=50
        rate_per_km=12
        if peak_hr=="y":
            fare=base_fare+(1.25*rate_per_km*distance)
            print("Fare :",fare,"Rs.")
            flag=False
        else:
            fare=base_fare+(rate_per_km*distance)
            print("Fare :",fare,"Rs.")
            flag=False

calc_fare()
```

Problem 5: Employee Attendance Eligibility Checker  
  
Problem Statement:   
Create a function that:  
Accepts attendance list ("P", "A")  
Calculates attendance percentage  
Returns Eligible if ≥ 75% else Not Eligible  
  
Real-Time Use:  
HR payroll systems   
Internship tracking  

```python
def eligibility(attendence):
    present=0
    for att in attendence:
        if att=="P":
            present+=1
    present_per=(present/len(attendence))*100
    if present_per>=75:
        return "Eligible."
    return "Not Eligible."

attendence=["P","P","P","P","A","P","P"]
eligibility(attendence)
```

Problem 6: Password Strength Checker

Problem Statement:
Create a function to check password strength.

Rules:
Minimum length = 8
Must contain at least one digit
Must contain at least one special character (@#$)

Real-Time Use:
Login systems
Security validation


Hint:
Use loop over string
Use conditions inside function

```python
def check_pass():
    not_ok=True
    hasdigit=False
    haschar=False
    special_chars = "@#$"
    while(not_ok):
        password=input("Please Enter Password:")
        if(len(password)<8):
            print("Password must be at least 8 characters")
            print("Please retry.")
            continue
        for c in password:
            if c.isdigit():
                hasdigit=True
            if c in special_chars:
                haschar=True
        if not hasdigit:
            print("Password must contain at least one digit.")
            print("Please retry.")
            continue
        if not haschar:
            print("Password must contain at least one special character (@#$).")
            print("Please retry.")
            continue
        print("Password is Strong.")
        not_ok=False

check_pass()
```
